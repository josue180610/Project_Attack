```python
x = df_cancer.drop ("diagnosis", axis = 1)
y = df_cancer["diagnosis"].values
```

```python
# Vamos a partir la data para tener nuestra muestra de testeo y trainig

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 7, shuffle = True)
```

```python
# De acuerdo a la función aplicada, se tiene la data particionada para el testeo y trainig

print(X_train.shape)
print(X_test.shape)

# Siendo el tamaño de la data testeo: 455 registros y data training: 114 registros
```

```python
# Aplicamos el modelo logístico y su ajuste respectivo

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score # mide el número total de errores que ha generado un algoritmo
from sklearn.metrics import recall_score # mide la proporción de verdaderos positivos frente a falsos negativos
from sklearn.metrics import precision_score
```

```python
# Se solicitan sus métricas del modelo

target_pred = model.predict(X_test)

print(f'Recall Score: {recall_score(y_test, target_pred)}')
print(f'Accuracy Score: {accuracy_score(y_test, target_pred)}')
print(f'Precision Score: {precision_score(y_test, target_pred)}')

# El resultado del modelo indica una eficacia del 94.74%
```

```python
def val_accuracy(d_test, y_predict):

    accuracy = []
    true_value = d_test#d_test # valores verdaderos
    predictions = y_predict#y_predict # predicciones hechas por el modelo
    cont_true_val = 0

    #conteo de los valores que se predijeron correctamente
    for i in range (len(true_value)):
        if true_value[i] == predictions [i]:
            cont_true_val = cont_true_val + 1

    # calculamos el porcentaje
    result = cont_true_val / float(len(true_value)) * 100.0

    accuracy.append(result)

    #Accuracy
    print ('Accuracy %.3f%%' % (sum(accuracy)/float(len(accuracy))))
```

```python
# Se obtiene la métrica mediante la función creada en el punto anterior

val_accuracy(y_test, target_pred)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
```

```python
# Para una verificación del modelo se solicita la matriz de confusio

disp = ConfusionMatrixDisplay.from_estimator(
    model,
    X_test,
    y_test,
    display_labels=cancer_arr.target_names,
    cmap=plt.cm.Blues,
    normalize="true"
)
# El resultado más preponderante de la matriz es que existe un 15% de diagnósticos que se indican "Benignos",
# cuando en realidad son "Malignos".
```

```python
# Se aplica un segundo modelo SVC y su ajuste

from sklearn import svm
model_2=svm. SVC()
result = model_2.fit(X_train, y_train)
```

```python
# Se solicitan sus métricas del segundo modelo

target_pred_2 = model_2.predict(X_test)

print(f'Recall Score: {recall_score(y_test, target_pred_2)}')
print(f'Accuracy Score: {accuracy_score(y_test, target_pred_2)}')
print(f'Precision Score: {precision_score(y_test, target_pred_2)}')

# El resultado de este segundo modelo indica una eficacia del 94.74%
# Lo que nos podría sugerir que ambos modelos (logístico y SVC) son apropiados ya que tienen la misma eficacia
```
